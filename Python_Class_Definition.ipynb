In our case we only create one Class to work with, incorporating most of the previous functions in it. We must consider that as we analyze the data we may require to create new objects and define new functions.

```python
class MovieList:
    '''
    MovieList class retrieves the required data and includes most necessary methods to explore data.
    '''
    #Defining constructor.
    def __init__(self, length):
        '''
        Length must be an integer between 1 and 1000.
        It defines the length of dataset.
        '''
        self.length = length
        self.simple_list = movie_table(self.length)
        self.detailed_list = movie_list(self.length)
    
    #Defining methods.
    def display(self):
        '''
        Shows all dataset.
        '''
        x = self.detailed_list
        x = HTML(x.to_html(escape=False, formatters=dict(Poster=path_to_image_html)))
        return x
    
    def text_search(self):
        '''
        Search for some string in dataset.
        '''
        x = self.detailed_list
        x = x.drop('Poster', axis=1)
        x = interact(string_search, table=fixed(x), string='')
        return x
    
    def genre_search(self):
        '''
        Subsets according to movie genre.
        Some films include multiple genres.
        '''
        x = self.detailed_list
        x = x.drop('Poster', axis=1)
        y = search_list(x, 'Genre')
        x = interact(string_search, table=fixed(x), string=y)
        return x

    def director_search(self):
        '''
        Subsets according to movie director.
        Some films include multiple directors.
        '''
        x = self.detailed_list
        x = x.drop('Poster', axis=1)
        y = search_list(x, 'Director')
        x = interact(string_search, table=fixed(x), string=y)
        return x    
    
    def writer_search(self):
        '''
        Subsets according to movie writer.
        Some films include multiple writers.
        '''
        x = self.detailed_list
        x = x.drop('Poster', axis=1)
        y = search_list(x, 'Writer')
        x = interact(string_search, table=fixed(x), string=y)
        return x    

    def actor_search(self):
        '''
        Subsets according to specific actor.
        '''
        x = self.detailed_list
        x = x.drop('Poster', axis=1)
        y = search_list(x, 'Actors')
        x = interact(string_search, table=fixed(x), string=y)
        return x

    def language_search(self):
        '''
        Subsets according to movie language.
        Some films include multiple languages.
        '''
        x = self.detailed_list
        x = x.drop('Poster', axis=1)
        y = search_list(x, 'Language')
        x = interact(string_search, table=fixed(x), string=y)
        return x    
    
    def country_search(self):
        '''
        Subsets according to movie country.
        Some films include multiple countries appereances.
        '''
        x = self.detailed_list
        x = x.drop('Poster', axis=1)
        y = search_list(x, 'Country')
        x = interact(string_search, table=fixed(x), string=y)
        return x

    def rating_search(self):
        '''
        Subsets according to movie standarized rating.
        '''
        x = self.detailed_list
        x = std_rating(x)
        x = x.drop('Poster', axis=1)
        y = ['', 'General Audience', 'Parental Guidance Suggested', 'Parents Strongly Cautioned', 'Restricted', 'No Children 17 or Under', 'Not Rated']
        x = interact(string_search, table=fixed(x), string=y)
        return x
    
    def filter_search(self):
        '''
        Subsets according to multiple criterias.
        '''
        x = self.detailed_list
        x = search_engine(x)
        return x
    
    def recommendation(self):
        '''
        Subsets according to multiple criterias.
        Only shows movie title, poster and IMDb score.
        '''
        x = self.detailed_list
        x = recommendation_engine(x)
        return x
    
    def barplot(self):
        '''
        Graph barplot according to some criteria.
        '''
        x = self.detailed_list
        x = data_barplot(x)
        return x
    
    def histogram(self):
        '''
        Graph histogram according to some criteria.
        '''
        x = self.detailed_list
        x = data_histogram(x)
        return x
    
    def map_distribution(self):
        '''
        Map showing distribution.
        Total count may exceed lenght of dataset due to multiple country appeareances in some items.
        '''
        x = self.detailed_list
        x = geo_distribution(x)
        return x
    
    def boxplot(self):
        '''
        Graph boxplot according to some criteria.
        '''
        x = self.detailed_list
        x = data_boxplot(x)
        return x
```
